```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count=len(school_data_complete["school_name"].unique())

student_count=len(school_data_complete["student_name"])
student_count_f = float(student_count)
student_count_f="{:,.0f}".format(student_count_f)

budget=(school_data_complete["budget"].unique()).sum()
budget="${:,.2f}".format(budget)

ave_math=round(school_data_complete["maths_score"].mean(),6)

ave_read=round(school_data_complete["reading_score"].mean(),6)

math_pass=school_data_complete.loc[school_data_complete["maths_score"]>=50]
percent_math=round((len(math_pass)/student_count)*100,6)

read_pass=school_data_complete.loc[school_data_complete["reading_score"]>=50]
percent_read=round((len(read_pass)/student_count)*100,6)

overall_pass=school_data_complete.loc[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]
percent_overall=round((len(overall_pass)/student_count)*100,6)

overall_df=pd.DataFrame({"Total Schools": school_count,
                        "Total Students": student_count_f,
                        "Total Budget": budget,
                        "Average Maths Score": ave_math,
                        "Average Reading Score": ave_read,
                        "% Passing Maths": percent_math,
                        "% Passing Reading": percent_read,
                        "% Overall Passing": percent_overall}, index=[0])
overall_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
import numpy as np
def assign_pass(row):

    if row >= 50:
        result = 'PASS'
    else:
        result = np.nan
    return result
school_data_complete['maths_pass'] = school_data_complete['maths_score'].apply(assign_pass)
school_data_complete['reading_pass'] = school_data_complete['reading_score'].apply(assign_pass)

for i, r in school_data_complete.iterrows():
    if school_data_complete.at[i, 'maths_pass'] == school_data_complete.at[i, 'reading_pass'] == 'PASS':
        school_data_complete.at[i, 'overall_pass'] = 'PASS'
    else:
        school_data_complete.at[i, 'overall_pass'] = np.nan


school_name_group = school_data_complete.groupby(["school_name","type","budget"], as_index = False)[['maths_score','reading_score']].mean()
df1 = pd.DataFrame(school_name_group)
df1.columns = ['School Name','School Type','Total_School_Budget','Average Maths Score','Average Reading Score']
df1
school_students = school_data_complete.groupby(["school_name"], as_index = False)[['Student ID','maths_pass','reading_pass','overall_pass']].count()
df2 = pd.DataFrame(school_students)
df2.columns = ['school_name','Total_Students','passing_maths','passing_reading','passing_overall']
df2
df3 = df1.join(df2)
df3['Per Student Budget'] = df3.apply(lambda row: row.Total_School_Budget / row.Total_Students, axis = 1)
df3['% Passing Maths'] = df3.apply(lambda row: (row.passing_maths / row.Total_Students)*100, axis = 1)
df3['% Passing Reading'] = df3.apply(lambda row: (row.passing_reading / row.Total_Students)*100, axis = 1)
df3['% Overall Passing'] = df3.apply(lambda row: (row.passing_overall / row.Total_Students)*100, axis = 1)
df3  
df3.drop(columns=['passing_maths','passing_reading','passing_overall','school_name'])
school_com_df=df3[["School Name","School Type","Total_Students","Total_School_Budget",
                   "Per Student Budget","Average Maths Score","Average Reading Score",
                  "% Passing Maths","% Passing Reading","% Overall Passing"]]

school_com_df=school_com_df.set_index('School Name')
school_com_df = school_com_df.rename_axis(None)
school_com_df
```
